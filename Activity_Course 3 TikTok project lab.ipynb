### **Tugas 4. Menentukan pencilan**

Ketika membangun model prediktif, keberadaan pencilan dapat menjadi masalah. Misalnya, jika Anda mencoba memprediksi jumlah penayangan video tertentu, video dengan jumlah penayangan yang sangat tinggi dapat menimbulkan bias pada model. Selain itu, beberapa pencilan mungkin mengindikasikan masalah dengan cara pengambilan atau perekaman data.

Tujuan utama dari proyek TikTok adalah untuk membangun model yang memprediksi apakah sebuah video merupakan klaim atau opini. Analisis yang telah Anda lakukan menunjukkan bahwa tingkat keterlibatan video berkorelasi kuat dengan status klaimnya. Tidak ada alasan untuk percaya bahwa salah satu nilai dalam data TikTok ditangkap secara keliru, dan nilai tersebut sesuai dengan ekspektasi tentang bagaimana media sosial bekerja: sebagian kecil video mendapatkan tingkat keterlibatan yang sangat tinggi. Itulah sifat alami dari konten viral.

Meskipun demikian, ada baiknya Anda mengetahui berapa banyak titik data Anda yang dapat dianggap sebagai pencilan. Definisi pencilan dapat berubah berdasarkan detail proyek Anda, dan akan sangat membantu jika Anda memiliki keahlian dalam bidangnya untuk menentukan ambang batas. Anda telah mempelajari bahwa cara umum untuk menentukan pencilan dalam distribusi normal adalah dengan menghitung rentang interkuartil (IQR) dan menetapkan ambang batas 1,5 * IQR di atas kuartil ke-3.

Dalam dataset TikTok ini, nilai untuk variabel hitungan tidak terdistribusi secara normal. Nilai-nilai tersebut sangat condong ke kanan. Salah satu cara untuk memodifikasi ambang batas pencilan adalah dengan menghitung nilai **median** untuk setiap variabel dan kemudian menambahkan 1,5 * IQR. Ini menghasilkan ambang batas yang, dalam hal ini, jauh lebih rendah daripada jika Anda menggunakan kuartil ke-3.

Tulis perulangan for yang mengulang nama kolom dari setiap variabel hitungan. Untuk setiap iterasi:
1. Hitung IQR dari kolom tersebut
2. Hitung median dari kolom tersebut
3. Hitung ambang batas pencilan (median + 1,5 * IQR)
4. Hitung jumlah video dengan jumlah di kolom tersebut yang melebihi ambang batas pencilan
5. Cetak "Jumlah pencilan, {nama kolom}: {jumlah pencilan}"

```
Contoh:
Jumlah pencilan, jumlah_penayangan_video: ___
Jumlah pencilan, jumlah video_like_count: ___
Jumlah pencilan, jumlah video_share: ___
Jumlah pencilan, jumlah unduhan video: ___
Jumlah pencilan, jumlah komentar video: ___
```

````python
### YOUR CODE HERE ###
count_cools = ['video_view_count',
              'video_like_count',
              'video_share_count',
              'video_download_count',
              'video_comment_count']

for column in count_cools:
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    median = data[column].median()
    outlier_threshold = median + 1.5*iqr
    
    # count the number of values that exceed the outlier threshold
    outlier_count = (data[column]>outlier_threshold).sum()
    print(f'Number of outlier,{column}:', outlier_count)
````
